```python
import os
from glob import glob
```

```python
def rename_wav_files(directory_path):
    file_list = os.listdir(directory_path)
    # WAV 파일에 대해서만 처리
    wav_files = [file for file in file_list if file.endswith('.wav')]
    
    for idx, old_name in enumerate(wav_files, start=1):
        new_name = f"경적_{idx}.wav"
        old_path = os.path.join(directory_path, old_name)
        new_path = os.path.join(directory_path, new_name)
        os.rename(old_path, new_path)
        print(f"Renamed: {old_name} -> {new_name}")

# 디렉토리 경로 설정
directory_path = "./raw_data/validation" 
rename_wav_files(directory_path)
```
